Exercício 1. Escreva uma função que conta a frequência de ocorrência de cada
palavra em um texto (arquivo txt) e armazena tal quantidade em um dicionário, onde a
chave é a vogal considerada.


```python
class Frequencia():
    def __init__(self) -> None:
        self.vogais = "aeiou"

    def contar_frequencia_palavras_por_vogal(self,nome_arquivo):

        frequencia_palavras_por_vogal = {}
        for vogal in self.vogais:
            frequencia_palavras_por_vogal[vogal] = {}

        with open(nome_arquivo, 'r') as arquivo:
            for linha in arquivo:
                palavras = linha.split()
                for palavra in palavras:
                    palavra = palavra.strip(".,!?()[]")  # Remover pontuações
                    if palavra:
                        primeira_letra = palavra[0]
                        if primeira_letra in self.vogais:
                            vogal = primeira_letra.lower()  # Usar minúsculas para as vogais
                            if palavra in frequencia_palavras_por_vogal[vogal]:
                                frequencia_palavras_por_vogal[vogal][palavra] += 1
                            else:
                                frequencia_palavras_por_vogal[vogal][palavra] = 1
        for vogal, frequencia in frequencia_palavras_por_vogal.items():
            print(f"Vogal {vogal}:")
            for palavra, contagem in frequencia.items():
                print(f"{palavra}: {contagem}")
            print()

        return frequencia_palavras_por_vogal

nome_arquivo = "C:\\Users\\victo\\Downloads\\stopwords.txt"  # Substitua pelo nome do seu arquivo
objeto = Frequencia()
resultado = objeto.contar_frequencia_palavras_por_vogal(nome_arquivo)
```

Exercício 2. Escreva uma função que apaga do dicionário anterior, todas as palavras que sejam ‘stopwords’. Ver https://gist.github.com/alopes/5358189


```python
class Frequencia():
    def __init__(self) -> None:
        self.vogais = "aeiou"

    def contar_frequencia_palavras_por_vogal(self,nome_arquivo):

        self.frequencia_palavras_por_vogal = {}
        for vogal in self.vogais:
            self.frequencia_palavras_por_vogal[vogal] = {}

        with open(nome_arquivo, 'r') as arquivo:
            for linha in arquivo:
                palavras = linha.split()
                for palavra in palavras:
                    palavra = palavra.strip(".,!?()[]")  # Remover pontuações
                    if palavra:
                        primeira_letra = palavra[0]
                        if primeira_letra in self.vogais:
                            vogal = primeira_letra.lower()  # Usar minúsculas para as vogais
                            if palavra in self.frequencia_palavras_por_vogal[vogal]:
                                self.frequencia_palavras_por_vogal[vogal][palavra] += 1
                            else:
                                self.frequencia_palavras_por_vogal[vogal][palavra] = 1

        return self.frequencia_palavras_por_vogal
    
    def remove_stopwords(self):
        stopwords = ['um', 'uma', 'o', 'a', 'e', 'é', 'de', 'do', 'da', 'em', 'no', 'na', 'para', 'com']

        for vogal in self.frequencia_palavras_por_vogal:
            palavras_vogal = self.frequencia_palavras_por_vogal[vogal].copy()  # Criar uma cópia para evitar problemas ao deletar itens como índices pulados ou KeyError.
            for palavra in palavras_vogal:
                if palavra in stopwords:
                    del self.frequencia_palavras_por_vogal[vogal][palavra]

        print("=========Dicionário sem STOPWORDS==========")
        for vogal, frequencia in self.frequencia_palavras_por_vogal.items():
            print(f"Vogal {vogal}:")
            for palavra, contagem in frequencia.items():
                print(f"{palavra}: {contagem}")
            print()



nome_arquivo = "C:\\Users\\victo\\Downloads\\stopwords.txt"  # Substitua pelo nome do seu arquivo
objeto = Frequencia()
resultado = objeto.contar_frequencia_palavras_por_vogal(nome_arquivo)
resultado = objeto.remove_stopwords()
```

Exercício 3. Escreva um programa que lê duas notas de vários alunos e armazena tais notas em um dicionário, onde a chave é o nome do aluno. A entrada de dados deve terminar quando for lida uma string vazia como nome. Escreva uma função que retorna a média do aluno, dado seu nome.

```python
class Nota():
    def __init__(self,nome, notas: []*2) -> None:
        self.nome = nome 
        self.notas = notas


    def media(self):
        dic_media = {}
        dic_media[self.nome] = sum(self.notas)/len(self.notas)

        return dic_media
resultado = []
while True:

    nome = input("Digite o nome do aluno ou aperte enter para sair: ")
    if nome == '':
        break
    notas = input("Digite as notas do aluno separadas por espaço: ")

    vet_notas = list(map(float,notas.split()))

    aluno = Nota(nome, vet_notas)
    objeto = aluno.media()
    resultado.append(objeto)

for pessoa in resultado:
    for na, media in pessoa.items():
        print(f"Aluno(a) {na} tem a média {media}")
        print() 
```

```python
class Notas():
    def __init__(self, dic_notas):
        self.__dic_notas = dic_notas

    def calcular_media(self, nome_aluno):
        if nome_aluno in self.__dic_notas:
            notas = self.__dic_notas[nome_aluno]
            notas = [float(nota) for nota in notas]
            media = sum(notas) / len(notas)
            return media
        else:
            return "Nome não encontrado"

dic = {}
while True:
    nome = input("Digite o nome do aluno ou aperte enter para imprimir as notas: ")
    if nome == '':
        break
    notas = input("Digite as notas do aluno separadas por espaço: ").split()

    dic[nome] = notas

gerenciar = Notas(dic)

for pessoa in dic.keys():
    media_aluno = gerenciar.calcular_media(pessoa)
    print(f"Aluno(a) {pessoa} tem a média {media_aluno}")
    print()
```

Exercício 4. Uma pista de Kart permite 10 voltas para cada um de 6 corredores. Escreva um programa que leia todos os tempos em segundos e os guarde em um dicionário, onde a chave é o nome do corredor. Ao final diga de quem foi a melhor volta da prova e em que volta; e ainda a classificação final em ordem (1o o campeão). O campeão é o que tem a menor média de tempos.


```python
class Tempo():
    def __init__(self, dic_media):
        self.__dic_media = dic_media

    
    def media(self):
        dic_ordenado = {}
        for nome, tempo in self.__dic_media.items():
            media = sum(tempo)/len(tempo)
            dic_ordenado[nome]=  media
        dic_ordenado = sorted(dic_ordenado.items(), key=lambda item:item[1])

        return dic_ordenado
    
         
contador = 1
dic = {}
while contador <= 6:

   
    nome_corredor = input("Digite o nome do corredor %d: " % (contador))
    tempo = input("Digite os tempos separados por espaço: ")
    vet_tempo = list(map(int,tempo.split()))

    dic[nome_corredor] = vet_tempo
    contador +=1


chave_ganhador, tempo_ganhador = min(dic.items(), key= lambda item: item[1])
tempo_ganhador = min(tempo_ganhador)
ranking = Tempo(dic)
resultado = ranking.media()

contador = 1

print(f"A melhor volta foi do corredor {chave_ganhador} no tempo {tempo_ganhador}")

for chave, valor in resultado:
    print(f"O corredor {chave} teve a média de tempo {valor: .2f} e ficou em {contador}° lugar")
    contador +=1
```

```python
class Agenda():
    def __init__(self):
        self.dic = {}
    
    def incluir_novo_nome(self,nome,telefones):
        self.nome = nome
        self.dic[self.nome] = telefones
        return self.dic

    def incluir_telefone(self):
        nome = input("Para qual nome você deseja incluir um novo número ? ").lower()
        if nome in self.dic.keys():
                novo_numero = input(f"Digite o número que deseja incluir para {nome}: ")
                telefones = self.dic.get(nome,"").split(', ') #obter os  números já existentes
                telefones.append(novo_numero)
                self.dic[nome] = ', '.join(telefones)
        else:
            controle = input("Nome não cadastrado. Deseja incluí-lo ? Responda 'S' ou 'N': ").lower()
            if controle == 's':
                novo_nome = input("Qual nome irá inserir ? ")
                novo_tel = ("Qual o telefone da nova inclusão ? ")
                self.incluir_novo_nome(novo_nome, novo_tel)
         

    def excluir_telefone(self):
        nome = input("De qual nome deseja excluir um número? ").lower()
        if nome in self.dic.keys():
            numero_exclusao = input(f"Qual número deseja excluir para {nome} ? Os números disponíveis são: {self.dic[nome]} ")
            telefones = self.dic[nome].split(',')  # Separar os telefones
            if numero_exclusao in telefones:
                telefones.remove(numero_exclusao)
                if len(telefones) == 0:
                    del self.dic[nome]  # Se não houver mais telefones, exclua a entrada
                else:
                    self.dic[nome] = ','.join(telefones)  # Atualize os telefones na entrada
                print(f"Número {numero_exclusao} excluído com sucesso.")
            else:
                print(f"Número {numero_exclusao} não encontrado para {nome}.")
        else:
            print(f"Nome {nome} não encontrado na agenda.")

    def excluir_nome(self):
        nome = input("Qual nome deseja excluir ? ").lower()
        if nome in self.dic.keys():
            del self.dic[nome]

    def consultar_telefone(self):
        nome = input("Deseja consultar o número de quem ? ").lower()
        if nome in self.dic:
            print(self.dic[nome])



agenda = Agenda()

while True:
    nome = input("Qual nome deseja incluir ?  Se quiser sair do programa, aperte enter")
    if nome == '':
        break
    telefone = input("Qual telefone deseja incluir ? Se quiser sair do programa, aperte enter ")
    if telefone == '':
        break
    agenda.incluir_novo_nome(nome, telefone)

    inc_tele = input("Deseja incluir um novo telefone para um contato já existente ? Responda 'S' ou 'N': " )

    if inc_tele.lower() == 's':
        agenda.incluir_telefone()

    excluir = input("Deseja excluir um número da agenda ?  Responda 'S' ou 'N' ")

    if excluir.lower() == 's':
        agenda.excluir_telefone()

    excluir_n = input("Deseja excluir um nome da agenda ?  Responda 'S' ou 'N' ")
    if excluir_n.lower() == 's':
        agenda.excluir_nome()

    consultar = input("Deseja consultar um número da agenda ?  Responda 'S' ou 'N  ")
    if consultar.lower()== 's':
        agenda.consultar_telefone() 
```
